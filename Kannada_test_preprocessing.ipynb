Text classification (basic)

When to use?

Learning NLP
College assignments
Small text processing tasks

2. what is fit_transform and transform ? difference between both.
                                fit learns from data
transform applies what was learned

What does fit() do?

fit() learns parameters from data

Examples of what it learns:

Mean & standard deviation (StandardScaler)

Vocabulary (CountVectorizer / TF-IDF)

Min & max values (MinMaxScaler)

note: No data is changed by fit() alone.

What does transform() do?

transform() uses the learned parameters to convert data

Examples:

Scales values

Converts text to vectors

Encodes categories

note: It cannot work without fitting first.


What is fit_transform()?

fit_transform() = fit() + transform() in one step
Learns from data
Immediately transforms the same data
note: Used mostly on training data.

Simple Real-Life Example
Teacher analogy

fit() → Teacher learns grading rules

transform() → Applies rules to grade papers

fit_transform() → Learns rules and grades at the same time
